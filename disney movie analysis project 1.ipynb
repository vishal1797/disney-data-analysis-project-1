# movie ratings

```python
plt.figure(figsize=(10, 6))
plt.title('movie ratings')
sns.countplot(x='mpaa_rating', data=data)
plt.xlabel('rating')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='genre',data=data)
plt.title("movie geners")
```

# movie geners total gross

```python
plt.figure(figsize=(18,6))
sns.lineplot(x='genre',y='total_gross',data=data)
plt.title('movie gross at different geners')
```

# movie gross with genres

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.barplot(x='total_gross', y='genre',data=data)
plt.title('movies total gross with genres')
plt.subplot(2, 1, 2)
sns.barplot(x='inflation_adjusted_gross', y='genre',data=data)
plt.title('movies inflation gross with genres')
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(x='inflation_adjusted_gross',y='genre',data=data)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=45)
sns.boxplot(x='genre',y='inflation_adjusted_gross',data=data)
```

# disney movies total gross with time

```python
plt.figure(figsize=(20, 6))
plt.title('disney movies income')
plt.xticks(rotation=90)
plt.plot('total_gross',data=data)
```

```python
avg_rating = data.groupby('mpaa_rating').sum()
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1).set_title('gross per rating')
sns.barplot(x=avg_rating['total_gross'], y=avg_rating.index)
plt.subplot(2, 1, 2)
sns.barplot(x=avg_rating['inflation_adjusted_gross'], y=avg_rating.index)
```
